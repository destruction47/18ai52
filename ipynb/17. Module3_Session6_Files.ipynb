# **Opening Files with the open() Function**
To open a file with the open() function, you pass it a string path indicating
the file you want to open; it can be either an absolute or relative path. 

The open() function returns a File object.

Try it by creating a text file named hello.txt using Notepad or TextEdit.

Type Hello world! as the content of this text file and save it in your user
home folder. 

Then, if you’re using Windows, enter the following into the interactive shell:

```
helloFile = open('F:\\CiTech\\PythonProgramming\\Module-3\\hello.txt')
```

This command will open the file in “reading plaintext” mode, or read mode for short. When a file is opened in read mode, Python lets only to read data from the file; the file can’t be written or modified in any way. 

Read mode is the default mode for files that open in Python. 

But if you don’t want to rely on Python’s defaults, you can explicitly specify the mode by passing the string value 'r' as a second argument to **open()** method. 

So **open('/Users/asweigart/hello.txt', 'r')** and 

open('/Users/asweigart/hello.txt') do the same thing.

```
helloFile = open('F:\\CiTech\\PythonProgramming\\Module-3\\hello.txt', 'r')
```

The call to **open() returns a File object**. 

A File object represents a file on the computer; it is simply another type of value in Python, much like the lists and dictionaries you’re already familiar with. 

In the previous example, **File object** was **stored** the  in the **variable helloFile**. 

Now, whenever there is a need to read from or write to the file, it can be done by calling methods on the File object in **helloFile**.

# **Reading the Contents of Files**
Now that there is a File object available, reading from it can be started. 

To read the entire contents of a file as a **string value**, use the File object’s **read()** method.  

Let’s continue with the hello.txt File object that was stored in **helloFile** variable.

Enter the following into the interactive shell:

```
helloContent = helloFile.read()
print(helloContent)
```

The output of the program looks as follows:

['Hi this is file Manipulation Program\n', 'Here we are trying to open a File and Perform various operations on it.']


# **Writing to Files**
Python allows to write content to a file in a way similar to how the print()
function “writes” strings to the screen. 

But we **can’t write** to a file that has been **opened in read mode**. 

Instead, **open** it in “**write** plaintext” mode or **“append** plaintext” mode, or write mode and append mode for short.


**Write mode** will **overwrite the existing file** and start from scratch, just like when a variable’s value is overwritten with a new value. 

Pass **'w'** as the second argument to **open()** to open the file in **write mode**. 

Append mode, on the other hand, will **append text to the end of the existing file** i.e. it keeps the pre existing content of the file intact. 

This can be thought of appending to a list in a variable, rather than overwriting the variable altogether. 

Pass **'a'** as the second argument to **open()** to open the file in **append** mode.


If the **filename passed to open() does not exist**, both **write** and **append** mode **will create a new, blank file**. 

**Important Note**: After **reading or writing a file**, call the **close() method** before opening the file again.

Enter the following into the interactive shell:

```
# -*- coding: utf-8 -*-
"""
Created on Fri Nov  6 14:46:31 2020

@author: Dr. K Satyanarayan Reddy

"""
baconFile = open('F:\\CiTech\\PythonProgramming\\Module-3\\bacon.txt', 'w')

#The above line creates a file called bacon.txt in write mode

baconFile.write('This line is being writen on to file called bacon.txt Hello world!\n')

baconFile.close()

baconFile = open('F:\\CiTech\\PythonProgramming\\Module-3\\bacon.txt', 'a')

baconFile.write('Bacon is not a vegetarian meal.')

baconFile.close()

baconFile = open('F:\\CiTech\\PythonProgramming\\Module-3\\bacon.txt')

content = baconFile.read()

baconFile.close()

print(content)
```

# **Saving Variables with the shelve Module**
Variables in the Python programs can be saved  to **binary shelf files** using
the **shelve** module. 

This way, the program can **restore data to variables from the hard drive**. 

The shelve module will allow to add Save and Open features to the program. 

**For example**: If a program is run  and entered some configuration settings, then  those settings can be savd to a shelf file and it will have the program load them the next time it is run.

Enter the following into the interactive shell:

```
import shelve
shelfFile = shelve.open('mydata')
cats = ['Zophie', 'Poola', 'Simon']
shelfFile['cats'] = cats
shelfFile.close()
```

**Note:** After running the previous code on Windows, it will be seen that  **three new files** in the **current working directory**: ***mydata.bak***, ***mydata.dat***, and ***mydata.dir*** are created by the Shelf module.

**These binary files contain** the **data** that was **stored in the shelf**. 

The format of these binary files is not important; we only need to know what 
the shelve module does, not how it does it. 

**Note**: The module frees one from worrying about how to store the program’s data to a file.

The programs can use the shelve module later, to reopen and retrieve the data from these shelf files. 

Shelf values don’t have to be opened in read or write mode—they can do both once opened. 

Enter the following into the interactive shell:

```
import shelve
shelfFile = shelve.open('F:\\CiTech\\PythonProgramming\\Module-3\\mydata')
type(shelfFile)
```

The **output** of this will be:

<class 'shelve.DbfilenameShelf'>

```
shelfFile['cats']
shelfFile.close()
```

The **output** of this will be: 

['Zophie', 'Poola', 'Simon']

Just like dictionaries, **shelf values have keys() and values()** methods that
will **return list-like values of the keys and values** in the shelf. 

Since these methods return list-like values instead of true lists, you should pass them to the list() function to get them in list form. 

Enter the following into the interactive shell:

```
import shelve

shelfFile = shelve.open('mydata')

print('\n', list(shelfFile.keys()))

print('\n', list(shelfFile.values()))

shelfFile.close()
```

The **output** of this will be:

['cats']

 [['Zophie', 'Poola', 'Simon']]

# **Saving Variables with the pprint.pformat() Function**
Recall that the **pprint.pprint()** function will “pretty print” the contents of a list or dictionary to the screen, while the **pprint.pformat()** function **will return this same text as a string** instead of printing it. 

Not only, **this string formatted is easy to read**, but it is also **syntactically correct Python code**. 

**Scenario**: Suppose there is a **dictionary stored in a variable** and this **variable has to be saved  and its contents** for future use. 

The **pprint.pformat()** will **give a string** that is to be  written to .py file. 

This file will be the module that can be imported whenever there is a need to use the variable stored in it.

**For example**, enter the following into the interactive shell:

```
import pprint

cats = [{'name': 'Zophie', 'description': 'chubby'}, {'name': 'Poola', 'description': 'fluffy'}, {'name': 'Dibba', 'description': 'skinny'} ]

pprint.pformat(cats)

fileObj = open('myCats.py', 'w')

fileObj.write('cats = ' + pprint.pformat(cats) + '\n')

fileObj.close()
```

Here, **pprint** is **imported**  for using the **pprint.pformat()**. 

Now there is a list of dictionaries, stored in a variable cats. 

To **keep the list in cats available even after we close the shell**, we use **pprint.pformat()** to return it as a string. 

Once we have the data in cats as a string, it’s easy to write the string to a file, which we’ll call **myCats.py**.

The modules that an import statement imports are themselves just Python scripts. 

When the **string from pprint.pformat() is saved** to a .py file, the file is a module that can be imported just like any other.

And since **Python scripts are themselves just text files** with the .py file
extension, so **the Python programs can even generate other Python programs**.


These files  can then be imported into scripts.

```
import myCats

myCats.cats

myCats.cats[0]

myCats.cats[0]['name']
```
