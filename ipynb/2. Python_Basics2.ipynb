A Slight Modification in the Program quoted above, where we can use Text matter 
with input() function....

```
# This program says hello and asks for my name.
print('Hello world!')
print('\nWhat is your name?') # ask for their name
myName = str(input('\nPlease Enter your name : ')) # Here we have used a Text line within input() function. 
print('\nIt is good to meet you, ' + myName)
```

***Dissecting The Program***: 
The **print()** Function:
The print() function displays the string value inside the parentheses on the
screen.
The line print('Hello world!') means “Print out the text in the string
'Hello world!'.” When Python executes this line, you say that Python is
calling the print() function and the string value is being passed to the function.
A value that is passed to a function call is an argument.

**The input() Function**
The input() function waits for the user to type some text on the keyboard
and press enter.
This function call evaluates to a string equal to the user’s text, and the
previous line of code assigns the myName variable to this string value.

**The len() Function**
You can pass the len() function a string value (or a variable containing a
string), and the function evaluates to the integer value of the number of
characters in that string.

```
myCollege = 'My College Name is Cambridge Institute of Technology'
print('The total length of the string contained in the string variable myCollege is = ', len(myCollege))
```

So you can see that intermediate spaces are also counted...

value from the integer or floating-point version, an integer can be equal to a
floating point.

```
55 == '55' # We are testing here whether integer value 55 is equal to string value 55
95 == 95.00 # We are testing here whether integer value 95 is equal to float value 55
```
