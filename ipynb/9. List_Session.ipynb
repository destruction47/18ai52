#**List-like Types**: Strings and Tuples

A string is similar to a “list” of single text characters

Similar to list, string supports *indexing*; *slicing*; and using them with  *for loops* , with *len()*, and with the *in* and *not in* operators

```
name = 'Zophie'
name[0]
```

```
name[-2]
```

```
name[0:4]
```

```
'Zo' in name
```

```
'z' in name
```

```
'p' not in name
```

```
for i in name:
  print('* * * ' + i + ' * * *')
```

List values are mutable(changeable) where as string is immutable (can't change)

```
name = 'Zophie a cat'
name[7] = 't'
```

The proper way to “mutate” a string is to use slicing and concatenation to build a new string by copying from parts of the old string


```
name = 'Zophie a cat'
newName = name[0:7] + 'the' + name[8:12]
newName
```

#The Tuple Data Type
Tuples are similar to String as they are also immutable, tuples cannot have their values modified, appended, or removed
Tuples are represented using (  ) common braces

```
eggs = ('hello', 42, 0.5)
eggs[0]
```

```
eggs[1:3]
```

```
len(eggs)
```

```
eggs = ('hello', 42, 0.5)
eggs[1] = 99
```

If we have only one value in the tuple, then, it can be specified by ',' comma operator

```
type(('hello',))
```

If we dont use ',' comma operator, then it consider it as string

```
type(('hello'))
```

Tuple is prefered over list as
1. Order of the values in tuple remain unchanged
2. As contents are immutable, python can apply code optimization for faster retrieval 

**Converting Types with the list() and tuple() Functions**
list() and tuple() are used to convert the given content to list type or tuple type respectively

```
tuple(['cat', 'dog', 5])
```

```
list(('cat', 'dog', 5))
```

```
list('hello')
```

Converting a tuple to a list is handy if you need a mutable version of a tuple value.

#**References**
A **reference** is a value that points to data and a **list reference** is a value that points to a list.

```
spam = [0, 1, 2, 3, 4, 5]
cheese = spam
cheese[1] = 'Hello!'
spam
```

```
cheese
```

The above example signifies that when a list is copied -> it only copy the reference, the list remains same.

**The copy Module’s copy() and deepcopy() Functions**

TO create a duplicate copy of the list/Dictionary

```
import copy
spam = ['A', 'B', 'C', 'D']
cheese = copy.copy(spam)
cheese[1] = 42
spam
```
