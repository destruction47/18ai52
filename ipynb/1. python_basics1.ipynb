# **Complete ORDER OF PRECEEDENCE**

`The order of operations (also called precedence) of Python math operators is similar to that of mathematics. The ** operator is evaluated first; the *, /, //, and % operators are evaluated next, from left to right; and the + and - operators are evaluated last (also from left to right). You can use parentheses to override the usual precedence if you need to. `

*NOTE: Parenthesis are always first in order LEFT to RIGHT.  
With Nested Parenthesi the Inner Parenthesis is evaluated First then Outer Parenthesis.*

image.png

```
(5 - 1) * ((7 + 1) / (3 - 1))
```

**Why Precedence is Required?**

Consider the evaluation of the Expression: 2 - 3 * 5
If we take (2 - 3) for evalution First then result will be -1 * 5 = -5

But if we take (3 * 5) for evalution First then result will be 2 - 15 = - 13

Type Conversion
You can convert from one type to another with the int(), float(), and complex() methods:

```
x = 1    # int
y = 2.8  # float
z = 1j   # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```
